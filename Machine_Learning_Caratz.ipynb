**7.2.1 Melakukan overview data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('jumlah-siswa-putus-sekolah-menurut-tingkat-tiap-provinsi-indonesia-sd-2023.csv')
data.head()
```

**7.2.2 Mengubah data dalam bentuk numerik**

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder

# Menyalin dataframe
data_int = data.copy()

# Inisialisasi LabelEncoder
encoder = LabelEncoder()

# Membuat list dari nama kolom data kategori
categorical_data = ['Provinsi', 'Tingkat - I', 'Tingkat - II', 'Tingkat - III', 'Tingkat - IV', 'Tingkat - V', 'Tingkat - VI', 'Jumlah', 'Status']

# Mengubah setiap kolom kategori menjadi numerik
for kolom in categorical_data:
  data_int[kolom] = encoder.fit_transform(data[kolom])

# Menampilkan dataframe yang telah diubah
data_int.head()
```

```python
import joblib

joblib.dump(encoder, 'encoder.pkl')
```

**7.2.3 Menganlisis data kategori**

```python
for kolom in categorical_data:
  print(kolom, data_int[kolom].unique())
```

```python
for kolom in categorical_data:
  print(kolom, data[kolom].unique())
```

**7.2.4 Menganalisis Matriks Korelasi**

```python
# Menampilkan matriks korelasi antar kolom
data_int.corr()
```

```python
plt.figure(figsize=(10,8))
plt.title('Korelasi Antar Kolom')
sns.heatmap(data_int.corr(), annot=True, linewidth=3)
plt.show()
```

**7.2.5 Distribusi Data**

```python
def distribusi():
  fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))
  plt.suptitle('Distribusi Data', fontsize=24)

  def kolom_generator():
    for kolom in data_int.select_dtypes(include=np.number):
      yield kolom
  kolom = kolom_generator()

  for i in range(0, 2):
    for j in range(0, 3):
      k = next(kolom)
      data_int[k].plot(kind='hist', ax=axes[i, j])
      axes[i, j].set_xlabel(k)

plt.show()
distribusi()
```

**7.2.6 Pemisahan Data**

```python
# Memisahkan data

data = data_int.drop('Jumlah', axis=1)
target = data_int['Jumlah']

# Memisahkan data untuk latihan dan tes
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.2)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

**7.2.7 Membuat Machine Learning Model**

```python
# Menggunakan linear SVC
from sklearn.svm import SVC

# Membuat objek nama model untuk memanggil SVC()
model = SVC(gamma='scale')
```

**7.2.8 Melatih Machine Learning Model**

```python
# Melatih model dengan data pelatihan (x_train dan y_train)
model.fit(x_train, y_train)
```

```python
# Membuat prediksi dengan data uji (x_test)
prediction = model.predict(x_test)

# Membuat visualisasi matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(y_test, prediction)
plt.xlabel('Negeri')
plt.ylabel('Swasta')
plt.title('Status Wilayah yang paling banyak tidak lulus atau putus sekolah')
plt.show()
```

**7.2.10 Analisa Confusion Matriks**

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Membuat confusion matrix
def display_conf(y_test,prediction):
    sns.heatmap(confusion_matrix(y_test, prediction), annot=True, linewidth=3, cbar=False)
    plt.title('Confusion Matriks')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

display_conf(y_test, prediction)
```

**7.2.10 Analisa eror matriks**

```python
from sklearn.metrics import r2_score, classification_report

print(f'R2 Score : {r2_score(y_test, prediction)}')
print('Classification_report :')
print(classification_report(y_test, prediction, zero_division=0))
```

**7.2.11 Meningkatkan Model**

```python
# Menggunakan GridSarchCV untuk menemukan model dengan parameter terbaik

from sklearn.model_selection import GridSearchCV
# Import StratifiedKFold untuk kontrol manual atas validasi silang
from sklearn.model_selection import StratifiedKFold

# Hyperparameter model SVC
param_grid = {'C':[0.01, 0.1, 1, 10, 100],
              'gamma':[100, 10, 1, 0.1, 0.01]}

# Buat objek StratifiedKFold dengan n_splits yang lebih rendah
# Pastikan n_splits kurang dari atau sama dengan jumlah minimum sampel di kelas mana pun
n_splits = min(5, y_train.value_counts().min())

# Ubah n_splits menjadi minimal 2 jika kurang dari 2
n_splits = max(2, n_splits)

# Membuat model terbaik dari semua kemungkinan kombinasi
best_model = GridSearchCV(SVC(), param_grid, cv=n_splits, refit=True)

# Melatih model terbaik
best_model.fit(x_train, y_train)
```

```python
# Model dengan parameter terbaik
best_model.best_estimator_
```

```python
# Membuat prediksi dengan model yang telah ditingkatkan
prediction = best_model.predict(x_test)

# Menampilkan confusion matriks pada prediksi yang baru
display_conf(y_test, prediction)
```

```python
print(f'R2 Score : {r2_score(y_test, prediction)}')
print('Classification_report :')
print(classification_report(y_test, prediction, zero_division=0))
```
